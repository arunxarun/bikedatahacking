# Bike Data Hacking

## Overview

Given the data coming from my gps/hr monitor fitness device, can I predict what my heart rate would be with any accuracy? Heart rate in this case is a suitable analog for effort. I'd like to train heart rate linear regression models on different sections of my training data to test for 'decoupling' which is when my heart rate is either lower or higher than expected. If it's lower, then I'm getting in shape. If it's higher, then I'm not. 

In other words, can I use a model to predict heart rate for a specific point in time, and can I compare that HR to the real HR to understand when the model might be based on old information -- my 'trailing' fitness?

## Creating Timestamps to Aggregate On.

In the snippet below I'm reading data parsed from tcx and dumped into a csv file into a data frame. Then I'm adding fields to that dataframe by rounding the timestamp of the row to 1, 5, 10 minutes. 

```python
# viz in notebook
%matplotlib inline 
import pandas as pd
from pandas import Series, DataFrame
from datetime import datetime
from dateutil.parser import parse
import numpy as np

colNames = ['str_ts','latDegrees','lngDegrees','altMeters','distMeters','hrBpm','speedMeters','type']
        
result = pd.read_table('./resources/activity_out.csv',sep='\t',header = None,names=colNames)

df = DataFrame(result)

# TODO: add timestamp value or convert string in line to manipulate timestamps.

df['ts'] = df['str_ts'].map(parse)
df.drop('str_ts',1,inplace=True)
df.set_index('ts',inplace=True)

def round_to_minute(ts,x=1):
    seconds = ts.hour*60*60 + ts.minute*60 + ts.second
    total_min = seconds / 60 # total minutes
    total_hours = total_min / 60 # hours
    remaining_mins = total_min % 60
    as_x_min_incs = (remaining_mins/x)*x
    new_t = datetime(ts.year,ts.month,ts.day,total_hours,as_x_min_incs)
    return new_t

def round_to_5minutes(ts):
    return round_to_minute(ts,5)

def round_to_10minutes(ts):
    return round_to_minute(ts,10)

df['minute'] = df.index.map(round_to_minute)
df['5minutes'] = df.index.map(round_to_5minutes)
df['10minutes'] = df.index.map(round_to_10minutes)

df[:10]
```

## Are These Really The Droids We Are Looking For? 

This 'buckets' the data in a really nice way. But it buckets the data on arbitrary time boundaries. I'm not sure that makes a ton of sense. These datapoints are time series data points, and *for time series an important indicator of future behavior is past behavior*. The reason arbitrary bucketing on time windows may not be appropriate for the kind of forecasting I want to do is that  when I bucket values into 1,5, and 10 minute buckets, then *for any one row in the set, it's X minute average is calculated from future values unless it is the last row*. 

```python
bucketDF = DataFrame()
bucketDF['hrBpm'] = df['hrBpm']
bucketDF['1MinHrBpm'] = df1MinAvgs['hrBpm']
bucketDF['5MinHrBpm'] = df5MinAvgs['hrBpm']
bucketDF['10MinHrBpm'] = df10MinAvgs['hrBpm']

bucketDF.plot()
```

Nothing shows up for the 1, 5, 10 minute datasets! WTF? 

```python
bucketDF[:10]
```

OK. Aggregations only have values on the aggregate boundaries -- in this case, the average HR is only stored on the minute boundary. That makes sense. When I think about it, it makes sense to backfill the average from each minute boundary. I tried doing fillna(), but that really wasn't what I was looking for. I was looking for an interpolation, which fortunately dataframes have:

```python
bucketDF.interpolate().plot()
```

So when I want to get an interpolated set of values in place, I do the following:

```python
bucketDF.interpolate(inplace=True)
bucketDF[10:30]
```

You can see from above that the 1 minute bucket starts having valid (not NaN) values at the minute boundary, and the second intervals are interpolated. The 5 minute and 10 minute buckets will have the same kind of interpolation.

## Take2: *Trailing* values

How do we get the trailing averages of the last 1,5,10 minutes? Even if we could do that, what if there aren't enough values to 'fill those buckets'? 

pandas has the ability to apply computations to a defined window of samples. This doesn't map well to a specific time boundary if your data isn't sampled on recurring time boundaries. The Suunto data is sampled roughly every second, but there are greater than 1 second gaps. So the 60/300/600 samples below are 'best effort' attempts to get 1,5, and 10 minute rolling means. 

```python
# rolling_mean takes a dataframe and a window expresssed as a number of periods.
%gui
d1 = pd.rolling_mean(df['hrBpm'],60)
d5 = pd.rolling_mean(df['hrBpm'],300)
d10 = pd.rolling_mean(df['hrBpm'],600)
newDf = DataFrame()
newDf['hrBpm'] = df['hrBpm']
newDf['60backHrBpm'] = d1
newDf['300backHrBpm'] = d5
newDf['600backHrBpm'] = d10

```

## Normalizing the data

So...now I have the windowed dataset. Before going further, I need to normalize the data. 
I'm going to use mean normalization to do this. 

For every value X in a columnX, I'm going to subtract mean(X) from X, then divide by stddev(X):

*norm(X) = (X - mean(X))/stddev(X)*

Note that while we are normalizing hrBpm1MinuteMean, hrBpm5MinuteMean, hrBpm10MinuteMean, we are _not_ normalizing hrBpm because that is the value we are trying to predict. 


```python
# get the mean and stddev of each column
mean = dfRolling.mean()
stddev = dfRolling.std()

print "mean"
print "-----------"
print mean
print "-----------"
print "stddev"
print "-----------"
print stddev
print "-----------"
```

```python

dfRollingNormed = DataFrame()

#dfRollingNormed = dfRolling.map(normalize)

def normalize(num,meanVal,stdDevVal):
    return (float(num - meanVal)/stdDevVal)
    
def normalizeAltMeters(val):
    return normalize(val, mean['altMeters'],stddev['altMeters'])
    
def normalizeSpeedMeters(val):
    return normalize(val, mean['speedMeters'],stddev['speedMeters'])

def normalizeAltMeters1MinMean(val):
    return normalize(val,mean['altMeters1MinMean'],stddev['altMeters1MinMean'])

def normalizeSpeedMeters1MinMean(val):
    return normalize(val,mean['speedMeters1MinMean'],stddev['speedMeters1MinMean'])

def normalizeHrBpm1MinMean(val):
    return normalize(val,mean['hrBpm1MinMean'],stddev['hrBpm1MinMean'])

def normalizeAltMeters5MinMean(val):
    return normalize(val,mean['altMeters5MinMean'],stddev['altMeters5MinMean'])

def normalizeSpeedMeters5MinMean(val):
    return normalize(val,mean['speedMeters5MinMean'],stddev['speedMeters5MinMean'])

def normalizeHrBpm5MinMean(val):
    return normalize(val,mean['hrBpm5MinMean'],stddev['hrBpm5MinMean'])

def normalizeAltMeters10MinMean(val):
    return normalize(val,mean['hrBpm10MinMean'],stddev['hrBpm10MinMean'])

def normalizeSpeedMeters10MinMean(val):
    return normalize(val,mean['speedMeters10MinMean'],stddev['speedMeters10MinMean'])

def normalizeHrBpm10MinMean(val):
    return normalize(val,mean['hrBpm10MinMean'],stddev['hrBpm10MinMean'])


dfRollingNormed['altMeters'] = dfRolling['altMeters'].map(normalizeAltMeters)
dfRollingNormed['speedMeters'] = dfRolling['speedMeters'].map(normalizeSpeedMeters)
dfRollingNormed['hrBpm'] = dfRolling['hrBpm']

dfRollingNormed['altMeters1MinMean'] = dfRolling['altMeters1MinMean'].map(normalizeAltMeters1MinMean)
dfRollingNormed['speedMeters1MinMean'] = dfRolling['speedMeters1MinMean'].map(normalizeSpeedMeters1MinMean)
dfRollingNormed['hrBpm1MinMean'] = dfRolling['hrBpm1MinMean'].map(normalizeHrBpm1MinMean)

dfRollingNormed['altMeters5MinMean'] = dfRolling['altMeters5MinMean'].map(normalizeAltMeters5MinMean)
dfRollingNormed['speedMeters5MinMean'] = dfRolling['speedMeters5MinMean'].map(normalizeSpeedMeters5MinMean)
dfRollingNormed['hrBpm5MinMean'] = dfRolling['hrBpm5MinMean'].map(normalizeHrBpm5MinMean)

dfRollingNormed['altMeters10MinMean'] = dfRolling['altMeters10MinMean'].map(normalizeAltMeters10MinMean)
dfRollingNormed['speedMeters10MinMean'] = dfRolling['speedMeters10MinMean'].map(normalizeSpeedMeters10MinMean)
dfRollingNormed['hrBpm10MinMean'] = dfRolling['hrBpm10MinMean'].map(normalizeHrBpm10MinMean)

dfRollingNormed[600:640]
```
